```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
year = [2019,2020,2021,2022,2023]
india_population = [1.3,3.2,2.8,4.5,6]
US_population = [2.3,1.6,3,5,4]
```

## Line Plot

```python
fig,axes = plt.subplots() #intialise the canvas
axes.plot(year,india_population,label="India",color="red")
axes.plot(year,US_population,label="USA",color="green")
axes.set_xlabel("Year")
axes.set_ylabel("Population in Billion")
axes.set_title("India VS US population")
axes.legend()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,5))
ax1.plot(year,india_population,label="India",color="red")
ax2.plot(year,US_population,label="USA",color="green")
ax1.set_xlabel("Year")
ax1.set_ylabel("population in billion")

ax2.set_xlabel("Year")
ax2.set_ylabel("population in billion")
```

## Scatter Plot

```python
import pandas as pd
```

```python
df = pd.read_csv("bike-sharing-demand.csv")
```

```python
df.head(5)
```

```python
fig,axes = plt.subplots(figsize=(15,10))
axes.scatter(df[df['season']==1]['temp'],df[df['season']==1]['humidity'],label="Spring")
axes.scatter(df[df['season']==2]['temp'],df[df['season']==2]['humidity'],label="Summer")
axes.scatter(df[df['season']==3]['temp'],df[df['season']==3]['humidity'],label="Autumn")
axes.scatter(df[df['season']==4]['temp'],df[df['season']==4]['humidity'],label="Winter")
axes.set_xlabel("temp in Celcius")
axes.set_ylabel("Humidity")
axes.legend()
```
